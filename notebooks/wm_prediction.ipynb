```python
import tensorflow as tf
import numpy as np
import pandas as pd
import copy
from matplotlib import pyplot as plt
import seaborn as sns

sns.set(style="whitegrid", color_codes=True)
```

```python
df = pd.read_parquet("../data/world_model_prediction.parquet", engine="pyarrow")
print(df.describe())
print(df.info())
df.head()
```

```python
# confidence interval for all states
sns.relplot(x="step",
            y="value",
            hue="state_type",
            #hue="source",
            style="source",
            kind="line",
            #row="rollout",
            ci="sd",
            #err_style="bars",
            #units="rollout",
            #estimator=None,
            data=df
           )
```

```python
# prediction boxplots for cosinus only
sns.boxplot(x="step",
            y="value",
            hue="state_type",
            #hue="source",
            data=df)
```
